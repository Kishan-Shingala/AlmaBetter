```python
def yearly_visual(dataframe, year=2016):
    # (mb_df - monthwise bookings dataframe) analyze and visualize for the year 2016
    mb_df = pd.DataFrame(dataframe[dataframe['arrival_date_year'] == year].groupby('arrival_date_month').size())
    mb_df['%'] = round((mb_df / mb_df.sum()) * 100, 1)
    mb_df = mb_df.rename_axis('2016', axis=1).rename(columns={0: 'count'})
    print(mb_df, '\n')

    plt.figure(figsize=(8, 5))
    sns.barplot(mb_df['count'])
    plt.xticks(rotation=45)
    plt.ylabel('Number of bookings')

    plt.text(0.07, 0.95, '% share', transform=plt.gca().transAxes, ha='center', fontsize=11, color='black')
    for index, row in mb_df.iterrows():
        plt.annotate(f'{row["%"]}%', (index, row["count"]), xytext=(0, 3), textcoords='offset points', ha='center')

    plt.twinx()
    sns.lineplot(mb_df['%'], color='black', marker='o')
    plt.ylabel('% Contribution')

    plt.title(f'"Month-wise bookings of {year}"')
    plt.show()
```

```python
class Data:
    def __init__(self, dataframe):
        self.df = dataframe

        # (fdf - formatted dataframe) for visualization, covering all values
        self.fdf = self.df.copy()

        # (cr - column renaming) dictionary
        self.cr = {0: 'count'}

        # (gc - generated colors) set to store previously generated colors
        self.gc = set()

        # (dcls - descriptive columns) for visualization
        self.dcls = ['min', 'mean', '50%', '75%', 'max']
        self.line = '\n----------------------------------------------------------------------------------------------------------------------\n'


    # (cum_per - cumulative percentage)
    def cum_per(self, dataframe):
        cum_df = pd.DataFrame({'cum_%': dataframe['%'].cumsum()})
        cum_df = pd.concat([dataframe, cum_df], axis=1)
        return cum_df


    # (cus_ord - custom order) for categorical column for a better visualization
    def cus_ord(self, column):
        # (uv - unique values)
        uv = self.fdf[column].value_counts().index
        if 'Others' in uv:
            if type(uv[0]) != str:
                order = sorted(filter(lambda index: not isinstance(index, str), uv)) + ['Others']
            else:
                order = list(filter(lambda index: index != 'Others', uv)) + ['Others']
        else:
            order = [index for index in uv]
        self.fdf[column] = pd.Categorical(self.fdf[column], categories=order, ordered=True)


    # (uhc - unique hex colors)
    def uhc(self, num_colors):
        # (ghc - generate hex color) generate random HEX color codes
        def ghc():
            """Generate a random HEX color code."""
            return "#{:06x}".format(random.randint(0, 0xFFFFFF))

        # (uc - unique colors)
        uc = set()
        while len(uc) < num_colors:
            color = ghc()
            if color not in self.gc:
                uc.add(color)
                self.gc.add(color)
        return list(uc)


    # (swp - size with percentage)
    def swp(self, columns_list, visual=True):
        self.swp_dic = {}
        # (fdwp_dic - formatted data with percentage dictionary) for use in visualization
        self.fdwp_dic = {}

        for column in columns_list:
            # (cdf - count dataframe)
            cdf = pd.DataFrame(self.df[column].value_counts().rename_axis(None))
            cdf['%'] = round((cdf / len(self.df)) * 100, 1)
            cdf = cdf.rename_axis(columns=column).rename(columns=self.cr)
            self.swp_dic[column] = cdf

            # (ri - relevant indexes)
            ri = set(cdf[cdf['%'] >= 2].index)
            if len(set(cdf.index) - ri) < 2:
                ri = cdf.index

            for index in cdf.index:
                if index not in ri:
                    self.fdf[column].replace(index, 'Others', inplace=True)
            self.cus_ord(column)

            # (fcd - formatted count data)
            fcd = pd.DataFrame(self.fdf[column].value_counts().rename_axis(None))
            fcd['%'] = round((fcd / len(self.df)) * 100, 1)
            self.fdwp_dic[column] = fcd.rename_axis(columns=column)

            if visual:
                print(self.cum_per(cdf).rename_axis(columns=column).head(), '\n')

                if fcd.index.size < 5:
                    plt.figure(figsize=(3, 5))
                    plt.pie(fcd['count'], labels=fcd.index, colors=self.uhc(fcd.index.size), autopct='%1.1f%%', startangle=0)
                else:
                    plt.figure(figsize=((fcd.index.size*0.9), 4.5))
                    sns.barplot(fcd['count'], palette=self.uhc(fcd.index.size))
                    plt.xlabel('')
                    plt.text(0.90, 0.90, '% share', transform=plt.gca().transAxes, ha='center', fontsize=11, color='black')
                    for index, row in fcd.iterrows():
                        plt.annotate(f'{row["%"]}%', (index, row["count"]), xytext=(0, 3), textcoords='offset points', ha='center')

                plt.title(f'"{column}"')
                plt.show()
                print(self.line)
        return self.swp_dic


    # (trd - two_way relational data), (ci - column index)
    def trd(self, ci, columns_list, visual=True):
        self.trd_dic = {}

        # (ri - row index)
        for ri in columns_list:
            if ci != ri:
                # (gdf - grouped dataframe)
                gdf = self.df.groupby([ri, ci]).size().unstack()[self.swp_dic[ci].head().index]
                counts = self.swp_dic[ri]['count']
                gdf['total_%'] = counts
                gdf = round(gdf.div(counts, axis=0).mul(100), 1)
                gdf = pd.concat([gdf, self.swp_dic[ri]], axis=1).rename_axis(index=ri, columns=ci).sort_values(by='count', ascending=False)
                self.trd_dic[ri] = gdf

                if visual:
                    print(gdf.head(), '\n')

                    # (lci - list of column index)
                    lci = self.swp_dic[ci].head().index
                    # (pv - percentage values)
                    pv = self.fdf.groupby([ri, ci]).size().unstack()[lci].apply(lambda x: (x/x.sum())*100, axis=1).T.values.flatten().tolist()

                    plt.figure(figsize=(5 if len(pv)<4 else len(pv)*0.9, 4.5))
                    sns.countplot(data=self.fdf, x=ri, hue=ci, palette=self.uhc(len(lci)))
                    for p, label in zip(plt.gca().patches, pv):
                        plt.annotate(f'{label:.1f}%', (p.get_x()+p.get_width()/2, p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

                    plt.xlabel('')
                    plt.title(f'"{ri}"')
                    plt.show()
                    print(self.line)
        return self.trd_dic


    # (dmc - descriptive multiple columns)
    def dmc(self, columns_list, visual=True):
        # (ddf - descriptive dataframe)
        ddf = round(self.df[columns_list].describe().transpose(), 1)

        if visual:
            for column in columns_list:
                values = ddf.loc[column, self.dcls].to_string(header=False)
                sns.violinplot(self.df, x=column, palette=self.uhc(1))
                plt.text(0.88, 0.75, f"{values}", transform=plt.gca().transAxes, ha='center', fontsize=11)
                plt.xlabel('')
                plt.title(f'"{column}"')
                plt.show()
                print(self.line)
        return ddf[self.dcls]


    # (tdc - two_way descriptive columns)
    def tdc(self, dc, columns_list, visual=True):
        self.tdc_dic = {}

        for column in columns_list:
            if column != dc:
                # (tdd - two_way descriptive data)
                tdd = round(self.df.groupby(column)[dc].describe(), 1)
                tdd = pd.concat([tdd[self.dcls], self.swp_dic[column]], axis=1).rename_axis(index=column, columns=dc).sort_values(by='count', ascending=False)
                self.tdc_dic[column] = tdd

                if visual:
                    print(tdd.head(), '\n')

                    # (ivl - index value list)
                    ivl = self.fdf[column].unique().tolist()
                    # (ln- length)
                    ln = len(ivl) if np.nan not in ivl else len(ivl) -1

                    plt.figure(figsize=(5 if ln<4 else ln*1.5, 4.5))
                    sns.violinplot(self.fdf, y=dc, x=column, hue=column, legend=False, palette=self.uhc(ln))
                    plt.xlabel('')
                    plt.title(f'"{column}"')
                    plt.show()
                    print(self.line)
        return self.tdc_dic


    def three_variable_charts(self, num_column, hue_column, columns_list):
        for column in columns_list:
            sns.scatterplot(data=self.df, x=column, y=num_column, hue=hue_column, palette=['#2ca02c', '#a70b00'])
            plt.show()
            print(self.line)
```
